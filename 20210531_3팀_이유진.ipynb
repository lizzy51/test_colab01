```python
#편의점 재고 관리
d_list=[]
#물건이름:product, 개수:number
while True:                            #무한반복
    product=input('물건이름:')
    if product=='':                    #반복중단 조건:엔터('')를 입력하면 반복을 빠져나간다.
        break
    number=int(input('개수:'))
    d_list[product]=number
d_list
```

```python
#존재하는 물건 - 개수 추가, 아니면 딕셔너리 추가
#물건이름:name, 개수:number
d_li={}
while True:                
    name=input('물건이름:')
    if name=='':         
        break
    number=int(input('개수:'))
    if name in d_li.keys():
        d_li[name]+=number
    else:
        d_li[name]=number
print(d_li)
#오름차순 정렬
for key,value in sorted(d_li.items()):
    print(key,value,end=' ')
```

```python
#집합set > 연산
sweety={'빼빼로','칸쵸','허니버터'}
salty={'허니버터', '포카칩', '새우깡'}

#합집합
print(sweety.union(salty))
print(sweety|salty)

#교집합
print(sweety.intersection(salty))
print(sweety&salty)

#차집합
print(sweety.difference(salty))
print(sweety-salty)
print(salty.difference(sweety))
print(salty-sweety)
```

```python
#예제_금액에 따른 할인율 계산하기
buy=0
buy=int(input('물건 구매가를 입력하세요:'))
if buy>=100000:
    dc=10
elif buy>=50000:
    dc=7.5
elif buy>=10000:
    dc=5
elif:
    dc=0

print(f'구매가:{buy}원')
print(f'할인율:{dc}%')
print(f'할인금액:{int(buy*dc/100}원')
print(f'지불금액:{int(buy-(buy*dc/100))}원')
```

```python
#예제_최소값 구하기
n1=int(input('숫자1:'))
n2=int(input('숫자2:'))
n3=int(input('숫자3:'))
min1=n1                   #첫번째 수 저장
if n2<min1:               #두번째와 비교
    min1=n2               #두번째 수가 작으면 min1 저장
if n3<min1:
    min1=n3
print('가장 작은 수:',min1)
```
