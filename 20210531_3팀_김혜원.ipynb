```python
#존재하는 물건 - 개수 추가, 아니면 딕셔너리 추가
d_li={}
#물건이름:name, 개수:num
while True:
    name=input('물건이름')
    if name =='':
        break
    num=int(input('num:'))
    if name in d_li.keys():
        d_li[name]+= num
    else:
        d_li[name]=num
print(d_li)
#오름차순정렬
for key,value in sorted(d_li.items()):
    print(key,value,end='')
```

```python
#집합연산
#합집합 a|b (원화기호)
#교집합a&b
#차집합a-b

li1 = {'빼빼로','칸초','허니버터'}
li2 ={'허니버터','포카칩','새우깡'}
#합집합
print(li1.union(li2))
print(li1|li2)
#교집합
print(li1.intersection(li2))
print(li1&li2)
#차집합
print(li1.difference(li2))
print(li1-li2)
print(li2.difference(li1))
print(li2-li1)
```
