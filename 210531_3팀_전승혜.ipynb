```python
#예제>편의점 재고 관리

store = {}
#물건이름 : product, 개수 : num
while True:
    product = input('물건이름: ')
    if product=='' :
        break
    num = int(input('개수: '))
    if product in store.keys():
        store[product] += num  #숫자 더하기
    else:
        store[product] = num
print(store)

#오름차순 정렬
for key, value in sorted(store.items()):
    print(key, value, end = ' ')
```

```python
#집합의 연산
# 합집합 a.union(b) a|b
# 교집합 a.intersection(b) a&b
# 차집합 a.difference(b) a-b

sweety = {'빼빼로', '칸쵸', '허니버터'}
salty = {'허니버터', '포카칩', '새우깡'}

#합집합
print(sweety.union(salty))
print(sweety|salty)
#교집합
print(sweety.intersection(salty))
print(sweety&salty)
#차집합
print(sweety.difference(salty))
print(sweety-salty)
print(salty.difference(sweety))
print(salty-sweety)
```

```python
#예제>금액에 따른 할인율 계산하기
price = int(input('물건 구매가를 입력하세요: '))

if price >= 100000:
    per = 0.1
elif price >= 50000:
    per = 0.075
elif price >= 10000:
    per = 0.05
else:
    per = 0   #구간에 없는 부분도 신경쓰자

dc = price*per
print(f'구매가: {price}원')
print(f'할인율: {per*100}%')
print(f'할인금액: {int(dc)}원')   #int를 씌워서 소수점 아래는 버리고 깔끔하게 나오기 위함
print(f'지불금액: {int(price-dc)}원')
```

```python
#예제>사칙연산 프로그램
num1 = float(input('숫자1 : '))
num2 = float(input('숫자2 : '))
operator = input('연산기호(+, -, *, /) : ')

if operator == '+':
    result = num1+num2
elif operator == '-':
    result = num1-num2
elif operator == '*':
    result = num1*num2
elif operator == '/':
    if num2 !=0:
        result = num1/num2
    else:
        result = None
print('-'*50)
print(f'{num1} {operator} {num2} = {result}')
```
