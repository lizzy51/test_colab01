```python
dict_p = {}
while True :
    a = input('이름 : ')
    if a == '':
        break
    b = int(input('개수 : '))
    if a in dict_p.keys():
        dict_p[a] += b
    else :
        dict_p[a] = b
print(dict_p)

for key,value in sorted(dict_p.items()):
    print(key,value,end=' ')
```

```python
## a와 b의 합집합 : a.union(b)   a|b
sweety = {'빼빼로','칸쵸','허니버터'}
salty = {'허니버터','포카칩','새우깡'}

print(sweety.union(salty))
print(sweety|salty)
```

```python
## a와 b의 교집합 : a.intersection(b)   a&b
print(sweety.intersection(salty))
print(sweety&salty)
```

```python
## a와 b의 차집합 : a.difference(b)   a-b
print(sweety.difference(salty))
print(sweety-salty)
```

```python
price = int(input('구매가(원) : '))
if price >= 100000:
    d = 0.1
elif price >= 50000:
    d = 0.075
elif price >= 10000:
    d = 0.05
else :
    d = 0
print(f'할인율 : {d*100}%')
print(f'지불금액(원) : {int(price*(1-d))}')
```

```python
num1 = int(input('숫자1 : '))
num2 = int(input('숫자2 : '))
cal = input('연산기호(+,-,*,/) : ')
if cal == '+':
    a = (num1 + num2)
elif cal == '-':
    a = (num1 - num2)
elif cal == '*':
    a = (num1 * num2)
elif num2 == 0:
    a = '무한대'
else:
    a = (num1 / num2)
print(f'{num1} {cal} {num2} = {a}')
```
